```
import requests
import json
page=1

# 裝產品資訊
save_data = []
while True:
    url=f'http://api.ichannels.com.tw/sitemember/main-goods.php?key=8FMPUF85PIzCD64mTv&member_code=af000121893&page={page}'
    response = requests.get(url)
    if response.json()['response_status']==1:
        save_data = save_data + response.json()['data']
        #把解析的json存檔
        print(f'讀取第{page}頁')
        print(f"共有資料{len(response.json()['data'])}筆\n")
    else:
        print("response_status:",response.json()['response_status'],"stop")
        break
    page=page+1

   
with open("data.json","a+",encoding="utf-8") as file:
    json.dump(save_data, file, ensure_ascii=False, indent=4)
# ensure_ascii 不要以ascii為儲存文字編碼
# indent 用漂亮模式保存資料，單位間隔為4個字元
print(f'總共{len(save_data)}筆資料')
```
